# Patterns in loading tables of data


## Using Spark.SQL
using the notebooks UI

```python
df = spark.sql("SELECT * FROM TheGreatLake.Batting_01 LIMIT 1000")
```

```python
df = spark.sql(f'''

SELECT * 
FROM TheGreatLake.Batting_01 LIMIT 1000

''')
```

## Using Spark.Read.Table
#### Loading Dataframe From Tables


```python
df2 = spark.read.table("TheGreatLake.Batting_01")
```

## Loading directly to Pandas Dataframe

This is for Data Wrangler

```python
pandasDF = spark.read.table("TheGreatLake.Batting_01").toPandas()
```

## Spark.Read.Format
Load using the absolute path of the table.



```python
df3 = spark.read.format('delta').load('abfss://4498e99d-670c-4d2b-b318-4d8747f21545@onelake.dfs.fabric.microsoft.com/44811cae-2fd7-4292-bd3f-a25b0bbda9ae/Tables/Batting_01')
```

Load using the relative path

```python
df4 = spark.read.format('delta').load('Tables/Batting_01')
```
