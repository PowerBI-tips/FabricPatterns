#### Load initial table using filter and aggregations

```python
df = spark.sql("SELECT yearID, sum(AB) as At_Bats, sum(H) as Hits FROM TheGreatLake.Batting_01 GROUP BY 1")
display(df)
```

#### Write data into a table using saveAsTable function

> Note: this removes your capitalization in the table name

```python
df.write.format("delta").saveAsTable("year_batting_v10")
```

```python
TestAtBat = spark.sql("SELECT * FROM TheGreatLake.year_batting_v10")
display(TestAtBat)
```

#### Saving table using a fully qualified Lakehouse path

```python
delta_path = "abfss://4498e99d-670c-4d2b-b318-4d8747f21545@onelake.dfs.fabric.microsoft.com/44811cae-2fd7-4292-bd3f-a25b0bbda9ae/Tables/year_batting_v20"

df.write.format("delta").save(delta_path)
```

#### Short hand version of writing data into a a table name

```python
delta_path = "Tables/year_batting_v30"

df.write.format("delta").save(delta_path)
```
